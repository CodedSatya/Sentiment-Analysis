```python
df = pd.read_csv("E:\Projects\Python\Flask\Sentiment Analysis\data.csv")
# new_df = df.dropna()
df
```

```python
data = df[["name","description", "remarks"]]
data
# for i in data:
#   print(i)
```

```python
from textblob import TextBlob
score = []

for i in data.index:
  # print(TextBlob(i).sentiment.polarity)
  row = data.loc[i]
  blob = TextBlob(row["description"])
  score.append((row['name'],blob.sentiment.polarity, blob.sentiment.subjectivity))
print(score)
```

```python
sentiment = []
for i in score:
    if i[2] == 0:
      continue
    else:
      if -1 <= i[1] < 0:
        sentiment.append((i[0], "Negative"))
      elif i[1] == 0:
        sentiment.append((i[0], "Neutral"))
      elif 0 < i[1] <= 1:
        sentiment.append((i[0],  "Positive"))
sentiment
```
